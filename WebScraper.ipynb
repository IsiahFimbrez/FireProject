```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
```

```python
url="https://www.usclimatedata.com/climate/acampo/california/united-states/usca2123"
page = requests.get(url)
print(page)
soup =  BeautifulSoup(page.content,'html.parser')
print(soup)
```

```python
table1=soup.findAll("table",{"id":"monthly_table_one"})
table2=soup.findAll("table",{"id":"monthly_table_two"})
len(table1)
len(table2)
```
